### Feature con valori più ricorrenti nelle 50 migliori predizioni

Si va ora a cercare quali sono le caratteristiche più ricorrenti delle abitazioni nelle 50 migliori predizioni. 

Per farlo si scorrerranno tutte le feature e si cerca per quali di esse la frequenza dei valori è superiore alla soglia di 30. Quindi si cercano le caratteristiche in comune che almeno 30 delle 50 case hanno.

```python
#si salvano in un dizionario le frequenze
most_frequent = {}
for i in best_50_CB.columns:
    most_frequent[i] = best_50_CB[i].value_counts()
```

```python
# vengono ordinati in ordine decrescente i valori delle frequenze
for i in most_frequent:
    most_frequent[i].sort_values(ascending=False)
```

Si estraggono le colonne che hanno più elementi in comune tra le varie istanze.

```python
lista = []
for i in most_frequent:
    tmp = most_frequent[i].tolist()
    if tmp[0] >= 30:
        lista.append(i)
#si stampano le colonne dove si predice meglio
print(lista)
```
