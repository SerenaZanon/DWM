```python
def data_preparation(dataframe):
    columns_names = np.array(dataframe.columns)
    first_values = np.array(dataframe.iloc[0])
    return columns_names, first_values
```

Per rendere le feature categoriali utilizzabili dai modelli è necessario trasformarle. 

Si è scelta la tecnica 'One Hot-encoding':
- `num_or_str` è una funzione che restituisce un vettore la cui dimensione è pari al numero di features presenti nel dataset; a seconda che la feature sia numerica o categoricale tale funzione pone in corrispondenza del suo indice 'True' (numeriche) o 'False' (categoriali);
- `names` è una funzione che accetta in input un dataframe e una lista di indici. Questa funzione restituisce una lista contenente i nomi delle feature del dataframe corrispondenti agli indici presenti nella lista di input.
- `process_data` è la funzione che applica la tecnica del One Hot-encoding su tutto il dataframe fornito.

```python
# Funzione che pone a 'False' le features categoricali e a 'True' le features numeriche
def num_or_str():
    rmv= []
    for i in range(len(first_values)):
        if isinstance(first_values[i], str):
            rmv.append(False)
        else:
            rmv.append(True)
    return rmv
```

```python
print(num_or_str())
```

Ciò significa che la prima feature è numerica, la seconda e la terza sono categoriali, la quarta è numerica, ...

```python
def names(dataframe, lista):
    tmp = []
    for i in lista:
        tmp.append(dataframe.columns[i])
    return tmp
```

```python
from sklearn.preprocessing import OneHotEncoder

def process_data(dataframe):
    
    # Features numeriche
    is_numerical  = np.array(num_or_str())
    # 'numerical_names' è una lista che contiene i nomi delle feature numeriche
    numerical_names = names(dataframe, np.flatnonzero(is_numerical))
    print(numerical_names)
    All_enc = pd.DataFrame(dataframe, columns = numerical_names)
    
    # Features categoriali
    # 'categorical_names' è una lista che contiene i nomi delle feature categoriche
    categorical_names = names(dataframe, np.flatnonzero(is_numerical == False))
    print(categorical_names)
    oh = OneHotEncoder(sparse = False, handle_unknown = 'ignore')
    oh.fit(dataframe[categorical_names])
    
    encoded = oh.transform(dataframe[categorical_names])
    All_enc = All_enc.fillna(0.0)
    for i,col in enumerate(oh.get_feature_names_out()):
        All_enc[col] = encoded[:,i]
    All_enc.columns = All_enc.columns.astype(str)
    
    return All_enc
```

```python
All_enc = process_data(df)
```

Si esegue lo split del dataset solo ai fini di vedere la dimensione che assumerà il training-set e il test-set (anche nei successivi notebook).

```python
def split(dataframe):
    
    # Si toglie la colonna da predire
    X = dataframe.drop(columns=['Sale_Price'], axis =1)

    # Si isola la colonna da predire
    y = dataframe['Sale_Price']

    X_train_80, X_test, y_train_80, y_test = train_test_split(X, y, test_size = 0.20, random_state = 3)
    
    return X_train_80, X_test, y_train_80, y_test
```

Si è imposto un `random_state` in modo che non cambino il training-set e il test-set altrimenti la valutazione cambierebbe ogni volta che lo split viene eseguito.

Quando nei prossimi notebook si importeranno dei dataset si utilizzerà sempre la `random_state = 3` in modo che i modelli vengano costruiti e testati sempre sugli stessi insiemi di istanze.

```python
X_train_80, X_test, y_train_80, y_test = split(All_enc)
```
